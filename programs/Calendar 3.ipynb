```python
#create list of 2 tuples for month and the days in the month for each month
C1 = [('January', range(1, 31 + 1)), 
            ('February', range(1, 28 + 1)), 
            ('March', range(1, 31 + 1)), 
            ('April', range(1, 30 + 1)), 
            ('May', range(1, 31 + 1)),
            ('June', range(1, 30 + 1)),
            ('July', range(1, 31 + 1)),
            ('August', range(1, 31 + 1)),
            ('September', range(1, 30 + 1)),
            ('October', range(1, 31 + 1)),
            ('November', range(1, 30 + 1)),
            ('December', range(1, 31 + 1))]
```

```python
#create list for days of week 
dof_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
```

```python
def calendar ():
    year = int(input("enter year"))
    s_day = input("enter first day of week")
    sday = dof_week.index(s_day)
    if leap_y(year):
        calendar[1] = 'February', range(1, 29+1)
    
    for month, days in C1:
        print('{0} {1}'.format(month, year).center(50, "/"))
        print(''.join(['{0:<12}'.format(x) for x in dof_week]))
        print('{0:<12}'.format(' ')*sday, end='')
    
        for day in days:
            print('{0:<12}'.format(day), end='')
            sday += 1
            if sday == 7:
                print()
                sday = 0
        print('\n')

            
def leap_y(year):
    if year % 4 == 0:
        if year % 100 == 0:
            if year % 400 == 0:
                return True
            else:
                return False
```
